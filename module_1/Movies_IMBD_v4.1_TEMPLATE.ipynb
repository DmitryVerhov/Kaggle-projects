# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
'''To solve this exesise I've needed to use combinations,which I've assigned'''
cast_comb={}# Dictionary for combinatios

for x in data['cast_list']:# Looping cast_list, craeted above
    for y in list(combinations(x, 2)):#Search for combinatios
      if y in cast_comb:
        cast_comb[y]=cast_comb[y]+1#apprising the value
      else:
        cast_comb.setdefault(y,1)# Creating new position in cast_comb
#Sorting:
cust_freq=list(cast_comb.items())
cust_freq.sort(key=lambda i: i[1])
#display(cust_freq[-5:])
#Choosing from given answers:
a27=cust_freq[-2][0]
print(a27)
```
